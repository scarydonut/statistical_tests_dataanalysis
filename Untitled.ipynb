```python
def spearman_rank_correlation(x, y):
    if len(x) != len(y):
        raise ValueError("Both lists must have the same length")
    n = len(x)
    rank_x = [sorted(x).index(i) + 1 for i in x]
    rank_y = [sorted(y).index(i) + 1 for i in y]
    d = [rank_x[i] - rank_y[i] for i in range(n)]
    d_squared_sum = sum([d_i ** 2 for d_i in d])
    rho = 1 - (6 * d_squared_sum) / (n * (n**2 - 1))
    return rho
x = [52,53,42,60,45,41,37,38,25,27]
y = [65,68,43,38,77,48,35,30,25,50]
rho = spearman_rank_correlation(x, y)
print(f"Spearman's rank correlation coefficient: {rho}")
```

```python
def pearson_correlation(x, y):
    if len(x) != len(y):
        raise ValueError("Both lists must have the same length")
    n = len(x)
    sum_x = sum(x)
    sum_y = sum(y)
    sum_x_squared = sum(xi ** 2 for xi in x)
    sum_y_squared = sum(yi ** 2 for yi in y)
    sum_xy = sum(xi * yi for xi, yi in zip(x, y))
    numerator = n * sum_xy - sum_x * sum_y
    denominator = ((n * sum_x_squared - sum_x ** 2) * (n * sum_y_squared - sum_y ** 2)) ** 0.5
    if denominator == 0:
        return 0 
    return numerator / denominator
x = [52,53,42,60,45,41,37,38,25,27]
y = [65,68,43,38,77,48,35,30,25,50]
r = pearson_correlation(x, y)
print(f"Pearson's correlation coefficient: {r}")
```

```python
def kendall_tau(x, y):
    if len(x) != len(y):
        raise ValueError("Both lists must have the same length")
    n = len(x)
    C = 0  
    D = 0 
    for i in range(n):
        for j in range(i + 1, n):
            if (x[i] > x[j] and y[i] > y[j]) or (x[i] < x[j] and y[i] < y[j]):
                C += 1  # Concordant pair
            elif (x[i] > x[j] and y[i] < y[j]) or (x[i] < x[j] and y[i] > y[j]):
                D += 1  
    tau = (C - D) / (0.5 * n * (n - 1))
    return tau
x = [52,53,42,60,45,41,37,38,25,27]
y = [65,68,43,38,77,48,35,30,25,50]
tau = kendall_tau(x, y)
print(f"Kendall's Tau correlation coefficient: {tau}")
```

```python
import pandas as pd
import numpy as np
from scipy.stats import f
def f_test(sample1, sample2, alpha=0.05):
    var1 = np.var(sample1, ddof=1) 
    var2 = np.var(sample2, ddof=1)  
    f_statistic = var1 / var2 if var1 > var2 else var2 / var1 
    df1 = len(sample1) - 1 
    df2 = len(sample2) - 1  
    p_value = 1 - f.cdf(f_statistic, df1, df2) 
    reject_null = p_value < alpha  
    return f_statistic, p_value, reject_null
def load_data_from_csv(csv_file, column1, column2):
    df = pd.read_csv(csv_file)
    sample1 = df[column1].dropna().values  
    sample2 = df[column2].dropna().values  
    return sample1, sample2
csv_file = 'tested.csv'  
column1 = 'Age'
column2 = 'Fare'  
sample1, sample2 = load_data_from_csv(csv_file, column1, column2)
f_stat, p_value, reject_null = f_test(sample1, sample2)
print(f"F-statistic: {f_stat}")
print(f"P-value: {p_value}")
print(f"Reject the null hypothesis: {reject_null}")
if reject_null:
    print("The variances are significantly different.")
else:
    print("The variances are not significantly different.")
```

```python
import pandas as pd
import numpy as np
from scipy.stats import friedmanchisquare
def load_data_from_csv(csv_file, columns):
    df = pd.read_csv(csv_file)
    data = df[columns]
    return data
def perform_friedman_test(data):
    groups = [data[col].values for col in data.columns]
    statistic, p_value = friedmanchisquare(*groups) 
    return statistic, p_value
def interpret_results(statistic, p_value, alpha=0.05):
    print(f"Friedman test statistic: {statistic}")
    print(f"P-value: {p_value}")
    if p_value < alpha:
        print("Reject the null hypothesis: There is a significant difference between the groups.")
    else:
        print("Fail to reject the null hypothesis: There is no significant difference between the groups.")
csv_file = 'tested.csv'  
columns = ['Pclass', 'Age', 'Fare'] 
data = load_data_from_csv(csv_file, columns)
statistic, p_value = perform_friedman_test(data)
interpret_results(statistic, p_value)
```

```python
import pandas as pd
import numpy as np
from scipy.stats import rankdata
from scipy.stats import kruskal
def load_data_from_csv(csv_file, columns):
    df = pd.read_csv(csv_file)
    data = [df[col].dropna().values for col in columns]  
    return data
def perform_kruskal_wallis_test(data):
    h_statistic, p_value = kruskal(*data)
    return h_statistic, p_value
def nemenyi_test(data, alpha=0.05):
    total_n = sum([len(group) for group in data])
    k = len(data)
    all_data = np.concatenate(data)
    ranks = rankdata(all_data)
    group_ranks = []
    start_index = 0
    for group in data:
        end_index = start_index + len(group)
        group_ranks.append(ranks[start_index:end_index])
        start_index = end_index
    avg_ranks = [np.mean(ranks) for ranks in group_ranks]
    q_alpha = 1.96  
    cd = q_alpha * np.sqrt(k * (k + 1) / (6 * total_n))
    print(f"Critical Difference (CD): {cd}")
    print("Performing pairwise comparisons between groups:")
    for i in range(k):
        for j in range(i + 1, k):
            diff_rank = abs(avg_ranks[i] - avg_ranks[j])
            print(f"Group {i+1} vs Group {j+1} | Avg Rank Diff: {diff_rank}")
            if diff_rank > cd:
                print(f"-> Groups {i+1} and {j+1} are significantly different.")
            else:
                print(f"-> Groups {i+1} and {j+1} are not significantly different.")
def main():
    csv_file = 'tested.csv'
    columns = ['Survived', 'Age', 'Fare']  
    data = load_data_from_csv(csv_file, columns)
    h_statistic, p_value = perform_kruskal_wallis_test(data)
    print(f"Kruskal-Wallis H Statistic: {h_statistic}")
    print(f"P-value: {p_value}")
    if p_value < 0.05:
        print("\nThe Kruskal-Wallis test is significant. Proceeding with the Nemenyi test.")
        nemenyi_test(data)
    else:
        print("\nThe Kruskal-Wallis test is not significant. No need to proceed with the Nemenyi test.")
if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import numpy as np
from scipy.stats import rankdata, norm
from scipy.stats import kruskal
def load_data_from_csv(csv_file, columns):
    df = pd.read_csv(csv_file)
    data = [df[col].dropna().values for col in columns]  
    return data
def perform_kruskal_wallis_test(data):
    h_statistic, p_value = kruskal(*data)
    return h_statistic, p_value
def bonferroni_dunn_test(data, alpha=0.05):
    total_n = sum([len(group) for group in data])
    k = len(data)
    all_data = np.concatenate(data)
    ranks = rankdata(all_data)
    group_ranks = []
    start_index = 0
    for group in data:
        end_index = start_index + len(group)
        group_ranks.append(ranks[start_index:end_index])
        start_index = end_index
    avg_ranks = [np.mean(ranks) for ranks in group_ranks]
    num_comparisons = k * (k - 1) // 2
    adjusted_alpha = alpha / num_comparisons
    print(f"Adjusted alpha (Bonferroni): {adjusted_alpha}")
    print("Performing pairwise comparisons between groups:")
    for i in range(k):
        for j in range(i + 1, k):
            diff_rank = abs(avg_ranks[i] - avg_ranks[j])
            n_i = len(data[i])
            n_j = len(data[j])
            z = diff_rank / np.sqrt((total_n * (total_n + 1)) / (6 * (n_i + n_j)))
            p_value = 2 * (1 - norm.cdf(z))   
            print(f"Group {i+1} vs Group {j+1} | Z: {z:.4f} | P-value: {p_value:.4f}")
            if p_value < adjusted_alpha:
                print(f"-> Groups {i+1} and {j+1} are significantly different.")
            else:
                print(f"-> Groups {i+1} and {j+1} are not significantly different.")
def main():
    csv_file = 'tested.csv'
    columns = ['Pclass', 'Age', 'Fare']  
    data = load_data_from_csv(csv_file, columns)
    h_statistic, p_value = perform_kruskal_wallis_test(data)
    print(f"Kruskal-Wallis H Statistic: {h_statistic}")
    print(f"P-value: {p_value}")
    if p_value < 0.05:
        print("\nThe Kruskal-Wallis test is significant. Proceeding with the Bonferroni-Dunn test.")
        bonferroni_dunn_test(data)
    else:
        print("\nThe Kruskal-Wallis test is not significant. No need to proceed with the Bonferroni-Dunn test.")
if __name__ == "__main__":
    main()
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
def load_data(csv_file):
    data = pd.read_csv(csv_file)
    return data
def clean_data(data):
    data = data.dropna()
    data = data.loc[:, data.nunique() > 1]
    return data
def standardize_data(data):
    features = data.select_dtypes(include=[np.number])  # Only numerical columns
    scaler = StandardScaler()
    standardized_data = scaler.fit_transform(features)
    return standardized_data
def compute_pca_with_svd(data, n_components):
    U, S, VT = np.linalg.svd(data, full_matrices=False)
    U = U[:, :n_components]
    S = S[:n_components]
    transformed_data = U @ np.diag(S)
    return transformed_data, S, VT
def plot_explained_variance(S):
    explained_variance_ratio = (S ** 2) / (S ** 2).sum()
    cumulative_variance_ratio = np.cumsum(explained_variance_ratio)
    plt.figure(figsize=(8, 6))
    plt.plot(range(1, len(explained_variance_ratio) + 1), cumulative_variance_ratio, marker='o')
    plt.title("Cumulative Explained Variance by Principal Components")
    plt.xlabel("Number of Principal Components")
    plt.ylabel("Cumulative Explained Variance")
    plt.grid(True)
    plt.show()
def main():
    csv_file = 'tested.csv'  
    data = load_data(csv_file)
    data = clean_data(data)
    standardized_data = standardize_data(data)
    n_components = 2
    transformed_data, S, VT = compute_pca_with_svd(standardized_data, n_components)
    print("Transformed Data (Top 2 Principal Components):")
    print(transformed_data)
    plot_explained_variance(S)
    transformed_df = pd.DataFrame(transformed_data, columns=[f'PC{i+1}' for i in range(n_components)])
    transformed_df.to_csv("transformed_data.csv", index=False)
if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import numpy as np
from scipy import stats
df = pd.read_csv('tested.csv')
def one_sample_t_test(data, population_mean):
    t_stat, p_value = stats.ttest_1samp(data, population_mean)
    print("One-Sample t-test")
    print(f"T-statistic: {t_stat:.4f}, P-value: {p_value:.4f}")
    if p_value < 0.05:
        print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
    else:
        print("Fail to reject the null hypothesis: The sample mean is not significantly different from the population mean.")
    print()
def independent_sample_t_test(data1, data2):
    t_stat, p_value = stats.ttest_ind(data1, data2)
    print("Independent Sample t-test")
    print(f"T-statistic: {t_stat:.4f}, P-value: {p_value:.4f}")
    if p_value < 0.05:
        print("Reject the null hypothesis: The means of the two samples are significantly different.")
    else:
        print("Fail to reject the null hypothesis: The means of the two samples are not significantly different.")
    print()
def paired_t_test(data1, data2):
    t_stat, p_value = stats.ttest_rel(data1, data2)
    print("Paired Sample t-test")
    print(f"T-statistic: {t_stat:.4f}, P-value: {p_value:.4f}")
    if p_value < 0.05:
        print("Reject the null hypothesis: There is a significant difference between the paired samples.")
    else:
        print("Fail to reject the null hypothesis: There is no significant difference between the paired samples.")
    print()
one_sample_t_test(df['Survived'], 30)
independent_sample_t_test(df['Survived'], df['Pclass'])
paired_t_test(df['Survived'], df['SibSp'])
```
